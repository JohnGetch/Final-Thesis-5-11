```python
import sys
import stmpy
import numpy as np
import cancelv2 as cancel
from collections import deque
from decimal import *
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
'''
To run script:
From LV program:
    scan_freq = scanning frequency from RT using sample rate/oversampling
    fwd_ss = forward scan speed in m/s taken from "Speed Get"
    bwd_ss = backward scan speed in m/s taken from "Speed Get"
    img_w = pixels/line taken from "Scan Buffer Get"
    img_h = # of lines taken from "Scan Buffer Get"
    img_wm = image width taken from "Scan Frame Get"
    img_hm = image height taken from "Scan Frame Get"

python noise_image.py [scan_freq] [fwd_ss] [bwd_ss] [img_w] [img_h] [img_wm] [img_hm]
'''

# load data
i = 1
temp_data = np.loadtxt("Test Files/Data Logger/11-2-2018-scan-data" + "%(number)03d" % {"number": i} + ".dat", skiprows=7)
while True:
    i += 1
    try:
        temp_data = np.concatenate((temp_data, np.loadtxt("Test Files/Data Logger/11-2-2018-scan-data" + "%(number)03d" % {"number": i} + ".dat", skiprows=7)))
    except:
        break

x = temp_data[:,1]
y = temp_data[:,2]
z = temp_data[:,3]
```

```python
b = 5000
a = 1
plt.plot(x[:b])
plt.plot(x[-(b+a):-a])
# plt.plot(y)
plt.show()
```

```python
# scan parameters
scan_freq = 2000 # S/s
x_min = Decimal(x.min())
x_max = Decimal(x.max())
y_min = Decimal(y.min())
y_max = Decimal(y.max())
img_w = 256 # px/line
img_h = 256
img_wm = Decimal(20.000000E-9) # m/line
img_hm = Decimal(20.000000E-9)

xrange = x_max-x_min
y_range = y_max-y_min

pixel_dx = xrange / (img_w)
pixel_dy = xrange / (img_h)

dx = np.gradient(x)
```

```python
# empty arrays for output
fr_scan = np.empty((img_w,img_h), dtype=object)
bk_scan = np.empty((img_w,img_h), dtype=object)

for idx, el in enumerate(temp_data):
#     print(idx)
    ix = int((Decimal(el[1]) - x_min) // pixel_dx)
    iy = int((Decimal(el[2]) - y_min) // pixel_dy)
    if ix > 255 or iy > 255: print(ix, iy, el[1], el[2])
    if ix > 255: ix = 255
    if iy > 255: iy = 255
#     print(ix)
#     print(iy)
    if dx[idx] > 0:
#         print(fr_scan[ix,iy])
        if fr_scan[ix,iy] == None: fr_scan[ix,iy] = [el[3]]
        fr_scan[ix,iy].append(el[3])
#         fr_scan[ix,iy] = np.append(fr_scan[ix,iy], el[2])
    else:
#         print(bk_scan[ix,iy])
        if bk_scan[ix,iy] == None: bk_scan[ix,iy] = [el[3]]
        bk_scan[ix,iy].append(el[3])
#         bk_scan[ix,iy] = np.append(bk_scan[ix,iy], el[2])
```

```python
output=np.zeros((img_w,img_h))

for yi, row in enumerate(fr_scan):
    for xi, l in enumerate(row):
        if l != None: output[-xi,yi] = np.mean(np.array(l))
        else: output[-xi, yi] = np.mean(z)
            
output_rev=np.zeros((img_w,img_h))

for yi, row in enumerate(bk_scan):
    for xi, l in enumerate(row):
        if l != None: output_rev[-xi,yi] = np.mean(np.array(l))
        else: output_rev[-xi, yi] = np.mean(z)
```

```python
output=np.zeros((img_w,img_h))

for yi, row in enumerate(fr_scan):
    for xi, l in enumerate(row):
        if l != None: output[-xi,yi] = len(l)
        else: output[-xi, yi] = 0
            
output_rev=np.zeros((img_w,img_h))

for yi, row in enumerate(bk_scan):
    for xi, l in enumerate(row):
        if l != None: output_rev[-xi,yi] = len(l)
        else: output_rev[-xi, yi] = 0
```
