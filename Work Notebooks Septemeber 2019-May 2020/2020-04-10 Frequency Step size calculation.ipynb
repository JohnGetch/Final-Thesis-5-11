```python
plt.style.use('classic')
i = 50
e = 400
p = 200
s = linspace(0.001,2.001,2000)
pt = np.zeros(len(s))
for j,step in enumerate(s):
    print(e-i,end='\r')
    pt[j] = sum(p/linspace(i,e,int((e-i)/step)))
loc = (np.where(abs(pt/360 - 3) < 0.05))[0]

label1 = '# of Periods ='+str(p)
loglog(s,pt/360,label=label1)
loglog(s,3+0*s,label='3 hours')
label2 = str(around(s[loc[0]],3))+ ' Hz'
loglog(s[loc[0]],pt[loc[0]]/360,'o',label=label2)
tick_params(labelsize=20)
ylabel('Time (hours)',size=20)
xlabel('Frequency step size (Hz)',size = 20)
legend(loc=0,fontsize=20)
xlim(0.001,2.1)
```
